```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Генерируем распределение и строим график по нему

```python
rv = sts.chi2(7)
sample = rv.rvs(1000)
x = np.linspace(0, 25, 100)
pdf = rv.pdf(x)

plt.hist(sample, normed=True, label='histogram', bins=15)
plt.plot(x, pdf, label='theoretical pdf')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
```

### Генерируем выборки с различным N и сторим графики для них

```python
ns = [5, 10, 50]

for i, n in enumerate(ns):
    sample_means = np.array([rv.rvs(n).mean() for j in range(1000)])
    
    std = sample_means.std()
    mean = sample_means.mean()
    
    label = '$\sigma$={std:.3f}\n $\mu$ = {mean:.3f}'.format(std=std, mean=mean)
    plt.hist(sample_means, bins=30, label=label, range=[0, 14], normed=True)
    
    norm_rv = sts.norm(7, np.sqrt(2.*7./n))
    x = np.linspace(0, 14, 100)
    pdf = norm_rv.pdf(x)
    
    plt.plot(x, pdf, label='theoretical pdf')
    plt.title('n={n}'.format(n=n))
    plt.legend()
    plt.show()
```

### Вывод

При увеличении n, распределение средних значений начинает сходиться к 7, среднеквадратичное отклонение падает. Т.е. при увеличении n точность апроксимации растет.
