```python
from scipy.optimize import minimize, differential_evolution
from matplotlib import pyplot as plt
import numpy as np

f_x = lambda x: np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)

minimum_2 = minimize(f_x, 2, method='BFGS')
minimum_30 = minimize(f_x, 30, method='BFGS')
with open('submission_1.txt', 'w') as file:
    file.write('%.2f %.2f' % (minimum_2.fun, minimum_30.fun))
    
de_min = differential_evolution(f_x, [(1, 30)])
with open('submission_2.txt', 'w') as file:
    file.write('%.2f' % de_min.fun[0])

h_x = lambda x: f_x(x).astype(int)
minimum_30 = minimize(h_x, 30, method='BFGS')
de_min = differential_evolution(h_x, [(1, 30)])
with open('submission_3.txt', 'w') as file:
    file.write('%.2f %.2f' % (minimum_30.fun, de_min.fun))
```

### f(x) graph on the gap [1, 30]

```python
x = np.arange(0, 30, 0.1)
plt.plot(x, f_x(x))
plt.show()
```

### h(x) graph on the gap [1, 30]

```python
x = np.arange(0, 30, 0.01)
plt.plot(x, h_x(x))
plt.show()
```
